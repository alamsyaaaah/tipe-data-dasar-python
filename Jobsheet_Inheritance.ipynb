## Praktikum 1 - Inheritance Dasar

```python
print("=== PRAKTIKUM 1 ===")

class Kendaraan:
    def __init__(self, merk):
        self.merk = merk

    def mulai_mesin(self):
        print(f"Mesin kendaraan {self.merk} dinyalakan.")

class Mobil(Kendaraan):
    def __init__(self, merk, warna):
        self.merk = merk
        self.warna = warna

    def info_mobil(self):
        print(f"Mobil {self.merk} berwarna {self.warna}")

mobil = Mobil("Tesla", "Merah")
mobil.info_mobil()
mobil.mulai_mesin()
```

##Praktikum 02: Menggunakan super() dalam konstruktor

```python
print("=== PRAKTIKUM 2 ===")

class Person:
    def __init__(self, nama, usia):
        self.nama = nama
        self.usia = usia

    def perkenalkan_diri(self):
        print(f"Halo, saya {self.nama}, usia {self.usia}")

class Student(Person):
    def __init__(self, nama, usia, nim, jurusan):
        super().__init__(nama, usia)
        self.nim = nim
        self.jurusan = jurusan

    def perkenalkan_diri(self):
        super().perkenalkan_diri()
        print(f"NIM: {self.nim}, Jurusan: {self.jurusan}")

mhs = Student("Dewi", 20, "001", "Teknik Komputer")
mhs.perkenalkan_diri()
```

##Praktikum 03: Method Overriding dan super()

```python
print("=== PRAKTIKUM 3 ===")

class Hewan:
    def __init__(self, nama):
        self.nama = nama

    def bersuara(self):
        print("Suara hewan umum")

class Kucing(Hewan):
    def bersuara(self):
        print(f"{self.nama} berkata: Meow!")

h1 = Hewan("Makhluk")
k1 = Kucing("Kitty")

h1.bersuara()
k1.bersuara()
```

##Praktikum 04: Multi-level Inheritance (Pewarisan Bertingkat)

```python
print("=== PRAKTIKUM 4 ===")

class Organisme:
    def __init__(self, nama):
        self.nama = nama

    def bernapas(self):
        print(f"{self.nama} bernapas")

class Hewan(Organisme):
    def __init__(self, nama, makanan):
        super().__init__(nama)
        self.makanan = makanan

    def bergerak(self):
        print(f"{self.nama} bergerak")

class Mamalia(Hewan):
    def __init__(self, nama, makanan, kaki):
        super().__init__(nama, makanan)
        self.kaki = kaki

    def menyusui(self):
        print(f"{self.nama} menyusui")

kucing = Mamalia("Kucing", "Daging", 4)
kucing.bernapas()
kucing.bergerak()
kucing.menyusui()
```

##raktikum 05: Hierarchical Inheritance

```python
print("=== PRAKTIKUM 5 ===")

class Pegawai:
    def __init__(self, nama, gaji):
        self.nama = nama
        self.gaji = gaji

    def tampilkan(self):
        print(f"Nama: {self.nama}, Gaji: {self.gaji}")

class Manager(Pegawai):
    def tampilkan(self):
        print(f"[Manager] {self.nama}, Gaji: {self.gaji + 5000000}")

class Staff(Pegawai):
    def tampilkan(self):
        print(f"[Staff] {self.nama}, Gaji: {self.gaji + 1000000}")

m = Manager("Budi", 10000000)
s = Staff("Citra", 7000000)

m.tampilkan()
s.tampilkan()
```

##Praktikum 06: Memeriksa Tipe Instance dengan instance()

```python
print("=== PRAKTIKUM 6 ===")

class Bentuk:
    pass

class Persegi(Bentuk):
    pass

obj = Persegi()

print(isinstance(obj, Persegi))   # True
print(isinstance(obj, Bentuk))    # True
print(isinstance(obj, str))       # False
```

#Tugas siswa

```python
print("=== PROGRAM HIERARKI MAHASISWA ===\n")

# Kelas Induk
class Mahasiswa:
    def __init__(self, nama, nim, ipk):
        self.nama = nama
        self.nim = nim
        self.ipk = ipk

    def tampilkan_info(self):
        print(f"Nama: {self.nama}")
        print(f"NIM: {self.nim}")
        print(f"IPK: {self.ipk}")

    def hitung_predikat(self):
        if self.ipk >= 3.5:
            print("Predikat: Dengan Pujian")
        elif self.ipk >= 3.0:
            print("Predikat: Sangat Memuaskan")
        elif self.ipk >= 2.5:
            print("Predikat: Memuaskan")
        else:
            print("Predikat: Cukup")


# Kelas Anak 1
class MahasiswaSarjana(Mahasiswa):
    def __init__(self, nama, nim, ipk, semester, sks_lulus):
        super().__init__(nama, nim, ipk)
        self.semester = semester
        self.sks_lulus = sks_lulus

    def tampilkan_info(self):
        print("=== Mahasiswa Sarjana ===")
        super().tampilkan_info()
        print(f"Semester: {self.semester}")
        print(f"SKS Lulus: {self.sks_lulus}")


# Kelas Anak 2
class MahasiswaMagister(Mahasiswa):
    def __init__(self, nama, nim, ipk, judul_tesis, pembimbing):
        super().__init__(nama, nim, ipk)
        self.judul_tesis = judul_tesis
        self.pembimbing = pembimbing

    def tampilkan_info(self):
        print("=== Mahasiswa Magister ===")
        super().tampilkan_info()
        print(f"Judul Tesis: {self.judul_tesis}")
        print(f"Pembimbing: {self.pembimbing}")


# --- Program Utama ---
mhs1 = MahasiswaSarjana("Andi Saputra", "123456", 3.7, 6, 110)
mhs2 = MahasiswaMagister("Budi Santoso", "654321", 3.2, "AI untuk Pertanian", "Dr. Sari")

print("DATA MAHASISWA\n")

mhs1.tampilkan_info()
mhs1.hitung_predikat()

print("\n--------------------------\n")

mhs2.tampilkan_info()
mhs2.hitung_predikat()
```
